```python
from skimage.io import imread
import pyclesperanto_prototype as cle
import dask.array as da
import napari

from napari.types import ImageData
from skimage.filters import gaussian
from napari_time_slicer import time_slicer
import pyclesperanto_prototype as cle
```

```python

filename = r"E:\BiAPoL\Projects\PoL-BioImage-Analysis-TS-Early-Career-Track\data\CalibZAPWfixed_000154_max.tif"
original_image = imread(filename)

# add empty z axis at second position
original_image = original_image[:,None, :, :]

viewer = napari.Viewer()
image_layer = viewer.add_image(original_image)

napari.utils.nbscreenshot(viewer)
```

```python
@time_slicer
def skimage_gaussian(image: ImageData, sigma: int = 2, viewer: napari.Viewer = None) -> ImageData:
    print('called')
    return gaussian(image, sigma=sigma)

returned = skimage_gaussian(image_layer.data, sigma = 4,viewer = viewer)
cle.imshow(returned[99])
```
